```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

bdq = pd.read_csv("/media/jupiter/marina/project-msbt/bdq.txt", sep="\t", header=0)

dlm = pd.read_csv("/media/jupiter/marina/project-msbt/dlm.txt", sep="\t", header=0)
lnz = pd.read_csv("/media/jupiter/marina/project-msbt/lnz.txt", sep="\t", header=0)

print(bdq)
type(bdq)



```

```python
bdqplot = bdq.groupby("GENE").size().plot(kind="bar", stacked= "Tr", title="Number of Variants identified in genes \n for Bedquiline and Clofazimine resistance")
bdqplot.set_xlabel("The different genes")
bdqplot.set_ylabel("Frequency of the variants")
plt.tight_layout()
plt.savefig('/media/jupiter/marina/project-msbt/bdq_prevalence.png',pad_inches=0.2, dpi=300)




```
